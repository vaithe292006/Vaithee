```python
import matplotlib.pyplot as plt

#Sample data
x=[1,2,3,4,5]
y=[10,15,20,25,30]

# Plotting
plt.plot(x,y)

#Labels and title
plt.xlabel("x-axis (Time)")
plt.ylabel("y-axis (Value)")
plt.title("Simple Line Chart")

# Show the chart
plt.show()
```
